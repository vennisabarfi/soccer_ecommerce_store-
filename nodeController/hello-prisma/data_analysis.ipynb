```python
# import necessary packages
import pandas as pd
```

```python
from dotenv import load_dotenv
import os
```

```python

# Get the current working directory of the Jupyter notebook session
current_directory = %pwd

# Replace backslashes with forwards lashes to work with Pandas
current_directory = current_directory.replace('\\', '/')

# print(current_directory)  # Verify the formatted path
```

```python
# load datasets
file_path = f"{current_directory}/hello-prisma/"
boots_data = pd.read_csv(f"{file_path}footballbootsdb.csv")
```

```python
# keep only these columns
boots = boots_data[['BootsName', 'BootsMaterial','BootsBrand', 'BootsType', 'BootsPosition']].dropna()
# Rename columns 
boots = boots.rename(columns={
    'BootsName': 'bootsName',
    'BootsMaterial': 'bootsMaterial',
    'BootsBrand': 'bootsBrand',
    'BootsType': 'bootsType',
    'BootsPosition': 'bootsPosition'
})
boots.head(10)

```

```python
# write dataframe to csv
import os
os.makedirs(file_path, exist_ok=True) 
boots.to_csv(f'{file_path}boots.csv', index=False)
```

```python
#write dataframe to json
import os
os.makedirs(file_path, exist_ok=True) 
boots.to_json(f'{file_path}boots.json', lines=True, orient='records', indent=4)
```

```python
# Convert DataFrame to a list of dictionaries
records = boots.to_dict(orient='records')
print(records)

# Write the custom JSON file
import os
import json 
os.makedirs(file_path, exist_ok=True) 

with open(f'{file_path}boots.json', 'w') as f:
    f.write('[\n')
    for i, record in enumerate(records):
        json.dump(record, f, indent=4)
        if i < len(records) - 1:
            f.write(',\n')
        else:
            f.write('\n')
    f.write(']')
```
